```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset. Replace 'titanic.csv' with your own dataset file for your project.
# Ensure the dataset file exists and is in the same directory as this jupyter notebook.
e_v_p_df = pd.read_csv('Electric_Vehicle_Population_Data.csv')

# Display the first few rows of the dataframe to ensure it is loaded correctly
display(e_v_p_df.head())
```

```python
#  1. How many electric vehicles are registered each year?

ev_by_year = e_v_p_df["Model Year"].value_counts().sort_index()

plt.figure(figsize=(10,5))
sns.lineplot(x=ev_by_year.index, y=ev_by_year.values, marker="o", color="blue")
plt.title("Electric Vehciles Registered Each Year")
plt.xlabel("Model Year")
plt.ylabel("Number of EVs")
plt.show()
```

```python
# 2. Which electric vehicle brand are the most popular?

top_brands = e_v_p_df["Make"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_brands.values, y=top_brands.index, palette="viridis")
plt.title("Top 10 Electric Vehicle Brands")
plt.xlabel("Number of Vehicles")
plt.ylabel("Brand")
plt.show()
```

```python
# 3. Whcih citites have the most electric vehciles?

top_cities = e_v_p_df["City"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_cities.values, y=top_cities.index, palette="mako")
plt.title("Top 10 Cities with the Most Electric Vehciles")
plt.xlabel("Number of Vehicles")
plt.ylabel("City")
plt.show()
```

```python
# 4. Which counties have the most electric vehicles?

top_counties = e_v_p_df["County"].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.lineplot(x=top_counties.values, y=top_counties.index, palette="cubehelix")
plt.title("Top 10 Counties with most Electric Vehciles")
plt.xlabel("Number of Vehicles")
plt.ylabel("County")
plt.show()
```

```python
# 5. Are most electric vehicles battery only (BEV) or plug in hybrids (PHEV)?

ev_types = e_v_p_df['Electric Vehicle Type'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(ev_types, labels=ev_types.index, autopct="%1.1f%%", colors=["lightgreen","lightpink"])
plt.title("Electric Vehicle Types: BEV vs. PHEV")
plt.show()
```

```python
# 6.  What's the relationship between price and electric range?

plt.figure(figsize=(10,5))
sns.scatterplot(x="Electric Range",y="Base MSRP", data=e_v_p_df, alpha=0.5, color="orange")
plt.title("Relationship Between Price (MSRP) and Electric Range")
plt.xlabel("Electric Range (Miles)")
plt.ylabel("Price (USD)")
plt.show()
```

```python
#7. What is the averge driving range of electric vehicles?

avg_range = e_v_p_df["Electric Range"].mean()
print(f"Averge Driving Range: {avg_range:2f} miles")
```

```python
avg_range = e_v_p_df["Electric Range"].mean()

plt.figure(figsize=(5,5))
sns.barplot(x=["Average Range"], y=[avg_range], color="darkblue")
plt.title("Average driving Range of Electric Vehicles")
plt.ylabel("Miles")
plt.show()
```
